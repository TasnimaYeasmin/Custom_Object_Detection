```
import os
import random
import shutil
```

```
def split_data_set(image_dir, label_dir):

  name = os.path.split(image_dir)
  f_val = open('test.txt', 'a')
  f_train = open('train.txt', 'a')
  
  path, dirs, files = next(os.walk(image_dir))
  data_size = len(files)

  ind = 0
  data_test_size = int(0.1 * data_size)
  test_array = random.sample(range(data_size), k=data_test_size)
  
  for f in os.listdir(image_dir):
    if os.path.exists(os.path.join('train', f)) or os.path.exists(os.path.join('test', f)):
      continue

    if(f.split(".")[1] == "jpg"):
      ind += 1
      img_path = os.path.join(image_dir, f)
      label_path = os.path.join(label_dir, f.split(".")[0]+'.txt')

      if not os.path.exists(label_path):
        continue
      
      if ind in test_array:
        shutil.copy(img_path, "test/")
        shutil.copy(label_path, "test/")
        f_val.write('data/obj/'+f+'\n')
      else:
        shutil.copy(img_path, "train/")
        shutil.copy(label_path, "train/")
        f_train.write('data/obj/'+f+'\n')
  print("Done {}\n".format(name[1]))
```

```
f_val = open('test.txt', 'w')
f_train = open('train.txt', 'w')
list_dir = [['human_hand_images', 'human_hand_labels'], ['mug_images', 'mug_labels'], ['coffee_cup_images', 'coffee_cup_labels'], 
            ['flower_images', 'flower_labels'], ['pen_images', 'pen_labels']]

for i in list_dir:
  split_data_set(i[0], i[1])
```
